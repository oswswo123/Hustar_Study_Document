```python
#6.  X_train, y_train, X_test, y_test를 각각 구성하라.
# Train_set : Year 기준 2004년까지의 data.
# Test_set : Year 기준 2005년부터의 data.
# Hint: Use df[:'2004']
```

```python
#6. your code here
X_train = df[:"2004"].drop(labels="Direction", axis=1)
y_train = df[:"2004"]["Direction"]
y_train = y_train.replace("Up", 1)
y_train = y_train.replace("Down", 0)

X_test = df["2005":].drop(labels="Direction", axis=1)
y_test = df["2005":]["Direction"]
y_test = y_test.replace("Up", 1)
y_test = y_test.replace("Down", 0)
```

```python
#7. 각 X_train의 마지막 5개의 행과, X_test의 상위 5개의 행을 출력하라.
```

```python
#7. your code here

print(X_train.tail())
print(X_test.head())
```

```python
#8. LDA model를 정의하고 train data를 활용하여 학습하고 추정된 범주별 분포의 평균과 공분산을 출력하라.
```

```python
#8. your code here

lda = LinearDiscriminantAnalysis(store_covariance=True)
lda.fit(X_train, y_train)

print("means :", lda.means_)
print("covariance :", lda.covariance_)
```

```python
#9. train과 test set 각각에 대해 train accuracy와 test accuracy를 계산하라.
```

```python
#9. your code here

print("Train acc of LDA : ", accuracy_score(y_train, lda.predict(X_train)))
print("Test acc of LDA : ", accuracy_score(y_test, lda.predict(X_test)))
```

```python
#10. QDA model를 정의하고 train data를 활용하여 학습하고 추정된 범주별 분포의 평균을 출력하라.
```

```python
#10. your code here

qda = QuadraticDiscriminantAnalysis(store_covariance=True)
qda.fit(X_train, y_train)

print("means :", qda.means_)
print("covariance :", qda.covariance_)
```

```python
#11. train과 test set 각각에 대해 train accuracy와 test accuracy를 계산하라.
```

```python
#11. your code here

print("Train acc of QDA : ", accuracy_score(y_train, qda.predict(X_train)))
print("Test acc of QDA : ", accuracy_score(y_test, qda.predict(X_test)))
```

```python
#12. 5번의 Pairplot을 바탕으로 하나의 독립변수만을 선택하여 LDA 모델을 다시 학습하고, 그때의 accuracy를 출력하라.
```

```python
#12. your code here

lda_12 = LinearDiscriminantAnalysis()
lda_12.fit(X_train[["Today"]], y_train)

print("Train acc with single variable : ", accuracy_score(y_train, lda_12.predict(X_train[["Today"]])))
print("Test acc with single variable : ", accuracy_score(y_test, lda_12.predict(X_test[["Today"]])))
```

### 실습과제 2 : Vowel Speech Recognition 2

```python
#Recall the Vowel problem
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
import pandas as pd

np.warnings.filterwarnings('ignore')
%matplotlib inline

# set common plots properties and define commonly used colors
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['axes.linewidth'] = 0.5
BLUE, ORANGE, GRAY1, GRAY4 = '#57B5E8', '#E69E00', '#231F20', '#646369'
```

```python
#13. vowel data를 로드하여 변수명 df로 저장하고 반응변수의 범위를 0-10으로 수정하라.
#df의 상위 5개의 row를 출력하라.
```

```python
#13. your code here

df = pd.read_csv("./data/Vowel.txt")
df.head()
```

```python
#14. predictor를 리스트 형태로 변수명 features에 저장 후 출력하라.
features
```

```python
#14. your code here

features = [element for element in list(df.columns) if "x." in element]
features
```

```python
#15. is_train을 column을 활용하여 train과 test set을 구성하라.(test set: is_train == 0 )
#이를 바탕으로 X_train, y_train, X_test, y_test를 구성하라.
#X_train과 X_test의 개수를 출력하라.
```

```python
#15. your code here

# train_mask = (df["is_train"] == 1)
# test_mask = (df["is_train"] == 0)

# X = df.drop(labels="y", axis=1)
# X_train = X.loc[train_mask, :].drop(labels=["is_train", "row.names"], axis=1)
# X_test = X.loc[test_mask, :].drop(labels=["is_train", "row.names"], axis=1)

# Y = df[["y", "is_train"]]
# y_train = Y.loc[train_mask, :]["y"]
# y_test = Y.loc[test_mask, :]["y"]

df_train, df_test = df[df.is_train == 1], df[df.is_train == 0]
X_train, y_train = df_train[features].values, df_train.y.values
X_test, y_test = df_test[features].values, df_test.y.values

print("X_train:", len(X_train))
print("X_test:", len(X_test))
```

```python
#16. LDA 모델을 학습하고 accuracy를 계산하라. (use n_components = 2)
```

```python
#16. your code here

lda_vowel = LinearDiscriminantAnalysis(store_covariance=True, n_components=2)
lda_vowel.fit(X_train, y_train)

print("Train acc of LDA :", accuracy_score(y_train, lda_vowel.predict(X_train)))
print("Test acc of LDA :", accuracy_score(y_test, lda_vowel.predict(X_test)))
```

```python
#17. Test data에 대해 각 범주별 평균을 출력하라.
```

```python
#17. your code here

df_test[features + ["y"]].groupby(["y"]).mean()
```

```python
#18. Test data와 Test data의 범주를 활용하여 2차원 plot하라.
```

```python
#18. Complete the code below

GRAY4 = '#646369'

X_reduced = lda_vowel.transform(X_test)
means = lda_vowel.transform(lda_vowel.means_)

colors = np.array([
    '#000000', '#0000FF', '#A52A2A', '#A020F0', '#FF8C00', '#00FFFF',
    '#708090', '#FFEC8B', '#000000', '#FF0000', '#00FF00'])
fig, ax = plt.subplots(figsize=(6.5, 5.65), dpi=110)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], facecolors='none',
           edgecolors=colors[y_test-1], s=5, linewidth=0.6)
ax.scatter(means[:, 0], means[:, 1], facecolors='none', edgecolors=colors,
           s=5, linewidth=5)
for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(6)
ax.text(-4.89, 4.4, 'Linear Discriminant Analysis', color=GRAY4, fontsize=9)
ax.set_xlabel('Coordinate 1 for Test Data with True label', color=GRAY4, fontsize=8)
_ = ax.set_ylabel('Coordinate 2 for Test Data with True label', color=GRAY4, fontsize=8)
```

```python
#19. Test data를 통해 범주를 예측하고, 예측한 범주 값을 활용하여 Test data에 대한 각 범주별 평균을 출력하라.
```

```python
#19. your code here

y_pred = lda_vowel.predict(X_test)

temp_test = df_test
temp_test["y_pred"] = y_pred
temp_test[features + ["y_pred"]].groupby(["y_pred"]).mean()
```
