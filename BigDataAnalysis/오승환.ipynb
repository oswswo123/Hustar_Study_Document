```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn import metrics
```

### 실습 과제 1. 데이터 불러온 후 NaN 값이 있다면 해당 행 제거하고, train_data : test_data = 7:3 split
#### 종속변수는 PKT_Y, 나머지 변수들이 독립변수

```python
df = pd.read_csv("./data/contest.csv")

df.head()
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
# PKT_Y에 NULL값 많음. 제거

df = df.dropna(subset=["PKT_Y"], how="any")
df.shape
```

```python
from sklearn.model_selection import train_test_split

# 0   No.             2500 non-null   int64 
# 1   LOT_ID          2500 non-null   object 
# 2   WF_ID           2500 non-null   int64  
# 3   X_AXIS          2500 non-null   int64  
# 4   Y_AXIS          2500 non-null   int64  
# 5   GRADE_CD        2500 non-null   object 

X = df.drop(labels=["No.", "LOT_ID", "WF_ID", "X_AXIS", "Y_AXIS", "GRADE_CD", "PKT_Y"], axis=1)
Y = df["PKT_Y"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

### 실습 과제 2. 선형 커널, 3차항 커널, RBF 커널, 시그모이드 커널로 각각 SVM 모델 구현 후 accuracy, F1 score, confusion matrix, precision, recall (C=1, gamma = 0.1)

```python
from sklearn import svm

svm_linear = svm.SVC(kernel="linear", C=1, gamma=0.1)
svm_linear.fit(X_train, Y_train)
```

```python
svm_poly3d = svm.SVC(kernel="poly", degree=3, C=1, gamma=0.1)
svm_poly3d.fit(X_train, Y_train)
```

```python
svm_rbf = svm.SVC(kernel="rbf", C=1, gamma=0.1)
svm_rbf.fit(X_train, Y_train)
```

```python
svm_sigmoid = svm.SVC(kernel='sigmoid', C=1, gamma=0.1)
svm_sigmoid.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, f1_score, recall_score
```

```python
# linear kernel metrics

linear_predict = svm_linear.predict(X_test)

print("linear accuracy score :", accuracy_score(Y_test, linear_predict))
print("linear F1 score :", f1_score(Y_test, linear_predict))
print("linear precision score :", precision_score(Y_test, linear_predict))
print("linear recall score :", recall_score(Y_test, linear_predict))
print("linear confusion_matrix :\n", confusion_matrix(Y_test, linear_predict))
```

```python
# poly3d kernel metrics

poly3d_predict = svm_poly3d.predict(X_test)

print("poly3d accuracy :", accuracy_score(Y_test, poly3d_predict))
print("poly3d F1 score :", f1_score(Y_test, poly3d_predict))
print("poly3d precision score :", precision_score(Y_test, poly3d_predict))
print("poly3d recall score :", recall_score(Y_test, poly3d_predict))
print("poly3d confusion_matrix :\n", confusion_matrix(Y_test, poly3d_predict))
```

```python
# rbf kernel metrics

rbf_predict = svm_rbf.predict(X_test)

print("rbf accuracy :", accuracy_score(Y_test, rbf_predict))
print("rbf F1 score :", f1_score(Y_test, rbf_predict))
print("rbf precision score :", precision_score(Y_test, rbf_predict))
print("rbf recall score :", recall_score(Y_test, rbf_predict))
print("rbf confusion_matrix :\n", confusion_matrix(Y_test, rbf_predict))
```

```python
# sigmoid kernel metrics

sigmoid_predict = svm_sigmoid.predict(X_test)

print("sigmoid accuracy :", accuracy_score(Y_test, sigmoid_predict))
print("sigmoid F1 score :", f1_score(Y_test, sigmoid_predict))
print("sigmoid precision score :", precision_score(Y_test, sigmoid_predict))
print("sigmoid recall score :", recall_score(Y_test, sigmoid_predict))
print("sigmoid confusion_matrix :\n", confusion_matrix(Y_test, sigmoid_predict))
```

### 실습 과제 3. 하이퍼 파라미터 C 값(0.01, 0.1, 1, 10)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
hyperparameter_C_list = [0.01, 0.1, 1, 10]

for C_value in hyperparameter_C_list:
    svm_rbf_c = svm.SVC(kernel="rbf", C=C_value, gamma=0.1)
    svm_rbf_c.fit(X_train, Y_train)
    
    rbf_c_predict = svm_rbf_c.predict(X_test)
    
    print("-" * 30)
    print("C value : {}".format(C_value))
    print("accuracy :", accuracy_score(Y_test, rbf_c_predict))
    print("F1 score :", f1_score(Y_test, rbf_c_predict))
    print("-" * 30)
```

### 실습 과제 4. 하이퍼 파라미터 gamma 값(0.005,0.01,0.1, 1)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
hyperparameter_gamma_list = [0.005, 0.01, 0.1, 1]

for gamma_value in hyperparameter_gamma_list:
    svm_rbf_gamma = svm.SVC(kernel="rbf", C=1, gamma=gamma_value)
    svm_rbf_gamma.fit(X_train, Y_train)
    
    rbf_gamma_predict = svm_rbf_gamma.predict(X_test)
    
    print("-" * 30)
    print("gamma value : {}".format(gamma_value))
    print("accuracy :", accuracy_score(Y_test, rbf_gamma_predict))
    print("F1 score :", f1_score(Y_test, rbf_gamma_predict))
    print("-" * 30)
```

### 실습 과제 5. 하이퍼 파라미터 degree(1,2,3,4,5)에 따른 다항식 커널 SVM 모델 accuracy와 F1 score 변화 관찰

```python
hyperparameter_degree_list = [1, 2, 3, 4, 5]

for degree_value in hyperparameter_degree_list:
    svm_polynd = svm.SVC(kernel="poly", degree=degree_value, C=1, gamma=0.1)
    svm_polynd.fit(X_train, Y_train)
    
    polynd_predict = svm_polynd.predict(X_test)
    
    print("-" * 30)
    print("degree value : {}".format(degree_value))
    print("accuracy :", accuracy_score(Y_test, polynd_predict))
    print("F1 score :", f1_score(Y_test, polynd_predict))
    print("-" * 30)
```
