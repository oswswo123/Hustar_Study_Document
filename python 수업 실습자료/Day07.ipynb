### 상속 (Inheritance)

```python
class A:
    a = 1
    def print_a(self):
        print("A class")

class B(A):
    pass

aa = A()

bb = B()

aa.print_a()
bb.print_a()
```

### 메소드 오버라이딩(Method overriding)

```python
# super class의 __init__을 overriding

class A:
    def __init__(self):
        print("this is [A] class and [__init__] function")
        self.var1 = 0
        self.var2 = 0
        
class B(A):
    def __init__(self):
        super().__init__()
        print("this is [B] class and [__init__] function")
        self.var3 = 0
        self.var4 = 0

a = A()
print()
b = B()
```

### 연산자 오버로딩(Operator overloading)

```python
class BankAccount:
    def __init__(self, balance=0, name="none"):
        self.balance = balance
        self.name = name
    
    # 입금
    def deposit(self, amount): self.balance += amount
    
    # 출금
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            print("잔액 부족")
    
    # 계좌정보 출력
    def get_info(self):
        print("이름 : ", self.get_name())
        print("잔고 : ", self.get_balance())
    
    # + operator overloading
    def __add__(self, amount):
        self.deposit(amount)
        
    # += operator overloading
    def __iadd__(self, amount):
        self.deposit(amount)
        return self
    
    # getter
    def get_name(self): return self.name
    def get_balance(self): return self.balance

a = BankAccount(500, "Hong")
a + 200
a.get_info()
```

Python의 기본 자료구조 set를 모방한 클래스 Set을 구현하세요.
- 임의의 원소를 중복 없이, 순서 없이 담는 집합형 자료구조</br>
- 아래의 메서드/연산자를 명시된 기능대로 구현</br>
    생성자 (&#95;&#95;init&#95;&#95;)
        list를 받아 중복 제거. 매개 변수 없이 생성 시, 빈 집합 상태로 생성
    add(element)
        set에 element가 존재하지 않으면 추가
    discard(element)
        set에 element가 존재하면 삭제
    clear()
        set에 존재하는 모든 원소 삭제
    &#95;&#95;len&#95;&#95;()
        set에 존재하는 원소 개수 return
    &#95;&#95;str&#95;&#95;()
        set에 존재하는 원소를 "{1, 2, 3}"의 형태로 return
    &#95;&#95;contains&#95;&#95;(element) ---> in 멤버체크
        set에 element가 존재하면 True return, 아니면 False return
    self <= other ---> &#95;&#95;le&#95;&#95;(element)
        self가 other의 부분집합이면 True return, 아니면 False return
    self >= other ---> &#95;&#95;ge&#95;&#95;(element)
        other이 self의 부분집합이면 True return, 아니면 False return
    self | other ---> &#95;&#95;or&#95;&#95;(element)
        self와 other의 원소를 모두 포함하는 합집합 set return
    self & other ---> &#95;&#95;and&#95;&#95;(element)
        self와 other가 공통으로 포함하는 원소를 포함하는 교집합 set return
    self - other ---> &#95;&#95;sub&#95;&#95;(element)
        self의 원소 중 other에 없는 원소만을 포함하는 차집합 set return
    |=, &=, -= ---> &#95;&#95;ior&#95;&#95;(element), &#95;&#95;iand&#95;&#95;(element), &#95;&#95;isub&#95;&#95;(element)
        위의 기능에 맞추어 구현, In-place operation
        자신을 return 하는 것을 잊지 말 것
- list 관련 함수 사용 가능. 단, set method는 사용하면 안됨

```python
class Set:
    # list를 받아 중복 제거. 매개 변수 없이 생성 시, 빈 집합 상태로 생성
    def __init__(self, my_set = list()):
        self.my_set = list()
        for item in my_set:
            if item not in self.my_set:
                self.my_set.append(item)
    
    # set에 element가 존재하지 않으면 추가
    def add(self, element):
        if element not in self.my_set:
            self.my_set.append(element)
    
    # set에 element가 존재하면 삭제
    def discard(self, element):
        if element in self.my_set:
            self.my_set.remove(element)
    
    # set에 존재하는 모든 원소 삭제
    def clear(self):
        self.my_set.clear()
    
    # set에 존재하는 원소 개수 return
    def __len__(self):
        counter = 0
        for _ in self.my_set:
            counter += 1
        return counter
    
    # set에 존재하는 원소를 "{1, 2, 3}"의 형태로 return
    def __str__(self):
        temp_string = str()
        temp_string += "{"
        
        for item in self.my_set:
            temp_string += str(item)
            temp_string += ", "
            
        temp_string = temp_string[:-2]
        if not temp_string:
            temp_string += "{"
        temp_string += "}"
        
        return temp_string
    
    # set에 element가 존재하면 True return, 아니면 False return (in 멤버체크)
    def __contains__(self, element):
        if element in self.my_set:
            return True
        else:
            return False
    
    # self가 other의 부분집합이면 True return, 아니면 False return (self <= other)
    def __le__(self, other):
        for item in self.my_set:
            if item not in other.my_set:
                return False
        return True
    
    # other이 self의 부분집합이면 True return, 아니면 False return (self >= other)  
    def __ge__(self, other):
        for item in other.my_set:
            if item not in self.my_set:
                return False
        return True
    
    # self와 other의 원소를 모두 포함하는 합집합 set return (self | other)
    def __or__(self, other):
        tempset = Set()
        for item in self.my_set:
            tempset.add(item)
        for item in other.my_set:
            tempset.add(item)
        return tempset
    
    # self와 other가 공통으로 포함하는 원소를 포함하는 교집합 set return (self & other)
    def __and__(self, other):
        tempset = Set()
        for item in self.my_set:
            if item in other.my_set:
                tempset.add(item)
        return tempset

    # self의 원소 중 other에 없는 원소만을 포함하는 차집합 set return (self - other)
    def __sub__(self, other):
        tempset = Set(self.my_set)
        for item in other.my_set:
            tempset.discard(item)
        
        return tempset
    
    # 자신을 return 하는 것을 잊지 말 것 (|=)
    def __ior__(self, other):
        for item in other.my_set:
            self.add(item)
        return self

    # 자신을 return 하는 것을 잊지 말 것 (&=)
    def __iand__(self, other):
        tempset = self & other
        self.clear()
        for item in tempset.my_set:
            self.add(item)        
        return self
    
    # 자신을 return 하는 것을 잊지 말 것 (-=)
    def __isub__(self, other):
        for item in other.my_set:
            self.discard(item)
        return self
```

```python
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()
```
