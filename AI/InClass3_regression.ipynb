```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# exercise 1-1

x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1, 1)

plt.figure(figsize=(10, 8))
plt.title("Data", fontsize=15)
plt.xlabel("X", fontsize=15)
plt.plot(x, y, "ko")
```

```python
m = y.shape[0]

A = np.hstack([x**0, x])
A = np.asmatrix(A)

theta = (A.T*A).I*A.T*y

print("theta : \n", theta)
```

```python
plt.figure(figsize=(10, 8))
plt.title("Regression", fontsize=15)
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.plot(x, y, "ko", label="data")

xp = np.arange(0, 5, 0.01).reshape(-1, 1)
yp = theta[0, 0] + theta[1, 0]*xp

plt.plot(xp, yp, "r", linewidth=2, label="regression")
plt.legend(fontsize=15)
plt.axis("equal")
plt.grid(alpha=0.3)
plt.xlim([0, 5])
plt.show()
```

```python
# exercise 1-2

import cvxpy as cvx

n = 100
x = -5 + 15*np.random.rand(n, 1)
noise = 10*np.random.randn(n, 1)

y = 10 + 1*x + 2*x**2 + noise

A = np.hstack([x**0, x, x**2])
A = np.asmatrix(A)

theta2 = cvx.Variable([3, 1])
obj = cvx.Minimize(cvx.norm(A*theta2-y, 2))
cvx.Problem(obj, []).solve()

print("theta2 :\n", theta2.value)
```

```python
# exercise 2

n = 100            
x = -8 + 15*np.random.rand(n, 1)
noise = 100*np.random.randn(n, 1)
# 수식은 직접 입력 
y = -2*x**3 - 3*x**2 + 2*x - 8 + noise

A = np.hstack([x**0, x, x**2, x**3])
A = np.asmatrix(A)

theta = (A.T*A).I*A.T*y
print("theta : \n", theta)

xp = np.linspace(np.min(x), np.max(x))
yp = theta[0, 0] + theta[1, 0]*xp + theta[2, 0]*xp**2 + theta[3, 0]*xp**3

plt.figure(figsize=(10, 8))
plt.plot(x, y, "o", markersize=4, label="actual")
plt.plot(xp, yp, "r", linewidth=2, label="estimated")

plt.title("Nonlinear regression", fontsize=15)
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.xlim([np.min(x), np.max(x)])
plt.grid(alpha=0.3)
plt.legend(fontsize=15)
plt.show()
```

```python
# exercise 2-2

x = np.linspace(-4.5, 4.5, 15).reshape(-1, 1)
y = np.array([0.9819, 0.7973, 1.9737, 0.1838, 1.3180, -0.8361, -0.6591, -2.4701, -2.8122, -6.2512,
              0.0023, -2.5234, 0.2253, -1.5343, -4.5966]).reshape(-1, 1)

plt.figure(figsize=(13, 7))
plt.title('Data Points', fontsize = 20)
plt.plot(x, y, 'ko')
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.xlim(-5, 5)
plt.ylim(-7, 3)
plt.show()
```

```python
A = np.hstack([x**0, x, x**2, x**3, x**4, x**5])
A = np.asmatrix(A)

theta = (A.T*A).I*A.T*y
print("theta : \n", theta)

xp = np.linspace(np.min(x), np.max(x))
yp = theta[0, 0] + theta[1, 0]*xp + theta[2, 0]*xp**2 + theta[3, 0]*xp**3 + theta[4, 0]*xp**4 + theta[5, 0]*xp**5
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Data')
plt.plot(xp, yp, label='Curve')
plt.axis([-5, 5, -6, 3])
plt.title('Regression with Polynomial basis', fontsize=15)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.grid(alpha=0.3)
plt.legend(fontsize=15)
plt.show()
```
